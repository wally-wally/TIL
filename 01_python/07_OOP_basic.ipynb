````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.(ë‚´ê°€ í‘¼ ë°©ë²•)
class Stack:
    
    def __init__(self):
        self.items = []
    
    
    def empty(self):
        if self.items == []:
            return True
        else:
            return False
    
    
    def top(self):
        if self.items == []:
            return None
        else:
            return self.items[-1]
    
    
    def pop(self):
        if self.items == []:
            return None
        else:
            item = self.items[-1]
            self.items = self.items[:-1]
            return item
    
    
    def push(self, element):
        self.items += [element]
````

````python
class Circle:
    pi = 3.14
    
    def __init__(self, r, x=0, y=0):
        self.r = r
        self.x = x
        self.y = y
        
        
    def area(self):
        return self.r**2 * self.pi # Circle.pië¡œ ì¨ë„ ë™ì¼
    
    
    def circumference(self):
        return 2 * self.r * self.pi
    
    
    def center(self):
        return self.x, self.y
    
    
    def move(self, x, y):
        self.x = x
        self.y = y
        return self.x, self.y
````
