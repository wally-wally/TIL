# 행맨 게임 만들기 

## 답안 확인 하기

> 행맨 게임을 위한 코드를 작성해봅시다. 
>
> `isanswer(answer, letters)`는 정답 문자열(answer)과 추측 알파벳의 리스트(letters)를 받아서, True/False 결과를 반환합니다.
```python
isanswer('apple', ['a', 'p', 'l']) #=> False
isanswer('apple', ['a', 'p', 'l', 'e']) #=> True
```

````python
# 여기에 코드를 작성하세요.
````

````python
def isanswer(answer, letters):
    for indi_answer in answer:
        if indi_answer in letters:
            answer.strip(indi_answer)
        else:
            return False
    return True
````

````python
print(isanswer('apple', ['a', 'p', 'l']))
print(isanswer('apple', ['a', 'p', 'l', 'e']))
````

## 현재 상태 출력하기

> 행맨 게임을 위한 코드를 작성해봅시다. 
>
> `status(answer, letters)`는 정답 문자열(answer)과 추측 알파벳의 리스트(letters)를 받아서, 현재 결과를 반환합니다.
```python
status('apple', ['a', 'p', 'l']) # => 'appl_'
```

````python
# 여기에 코드를 작성하세요.
````

````python
def status(answer, letters):
    for i in range(len(answer)):
        if answer[i] not in letters:
            answer = answer.replace(answer[i], '_')
    return answer
````

````python
def status(answer, letters):
    return ''.join([answer.replace(answer[i], '_') for i in range(len(answer)) if answer[i] not in letters])
````

````python
status('apple', ['a', 'p', 'l'])
````

## 행맨 로직 만들기 - 자유코딩!

> 위의 코드를 활용하여 `hangman(answer)`를 만들어봅시다!
>
> 사용자가 답을 맞출 때까지 진행하며, 8번내로 정답을 맞춰야 합니다.


````python
# 자유코딩
````

````python
def hangman(answer):
    count = 0
    while count < 7:
        letter = input('알파벳을 입력하세요. : ')
        new_answer = status(answer, letter)
        count += 1
        print(f'시도 회수 : {count} / 결과 : {isanswer(answer, letter)} / 현재 상태 : {new_answer}')
        if isanswer(answer, letter):
            return 'You Win!'
    count += 1
    last_chance = input('마지막 기회! 정답은? :')
    if isanswer(answer, last_chance):
        return 'You Win!'
    else:
        return 'Game Over'
````

````python
hangman('apple')
````
