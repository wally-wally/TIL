# `abs()`

> 절대값은 숫자(int, float)가 들어오면 절대값을 반환하고, 복소수(complex)가 들어오면 그 크기를 반환합니다.
> 
> `my_abs(x)`를 만들어보세요.

```python
# 예제 입력 및 출력
print(abs(3+4j), abs(-0.0), abs(-5))
abs(0)
```

```python
# 아래에 코드를 작성해주세요.

# 내가 푼 방법

def my_abs(num):
    if type(num)==int or type(num)==float:
        if num >= 0:
            return num
        else:
            return -num
    else:
        return (num.real**2 + num.imag**2) ** (1/2)
```

```python
# 강사가 푼 방법

def my_abs(x):
    if type(x) == complex:
        return (x.real**2 + x.imag**2) ** (1/2)
    else:
        if x == 0:
            return x ** 2
        elif x < 0:
            return x * -1
        else:
            return x
```

```python
# 예제 입력 및 출력
print(my_abs(3+4j), my_abs(0.0), my_abs(-5))
```

# `all()`

> all은 인자로 받는 iterable(range, list)의 모든 요소가 참이거나 비어있으면 True를 반환합니다.
>
> 이와 같은 my_all(x)을 작성해보세요

```python
# 예제 입력 및 출력
print(all([1, 2, 5, '6']))
print(all([[], 2, 5, '6']))
```

```python
# 아래에 코드를 작성해주세요.

# 내가 푼 방법

def my_all(iter):
    count = 0
    if iter == []:
        return True
    for i in iter:
        if bool(i):
            count += 1
            if count == len(iter):
                return True
        else:
            return False
```

```python
# 강사가 푼 방법
# 모두 참이면 True + 빈 것도 True => 하나라도 False 라면 False

def my_all(elements):
    for element in elements:
        # element가 False이면 False 나온다.(element가 False일 경우 if문이 True가 되어야 함)
        if not element:
            return False
    return True
```

```python
# 예제 입력 및 출력
print(my_all([1, 2, 5, '6']))
print(my_all([[], 2, 5, '6']))
print(my_all([]))
```

# `any()`

> any는 인자로 받는 iterable(range, list)의 요소 중 하나라도 참이면 True를 반환하고, 비어있으면 False를 반환합니다.
>
> 이와 같은 my_any(x)를 작성해보세요.

```python
# 예제 입력 및 출력
print(any([1, 2, 5, '6']))
print(any([[], 2, 5, '6']))
print(any([0]))
```

```python
# 아래에 코드를 작성해주세요.
def my_any(iter):
    count = 0
    if iter == []:
        return False
    for i in iter:
        if not bool(i):
            count += 1
            if count == len(iter):
                return False
        else:
            return True
```

```python
# 강사가 푼 방법
# 어느 하나라도 True 이면 True + 빈 것은 False

def my_any(elements):
    for element in elements:
        # 만약 element가 True라면 True return
        if element: # == True 굳이 안 써도 됨
            return True
    return False
```

```python
# my_any 예제 입력 및 출력
print(my_any([1, 2, 5, '6']))
print(my_any([[], 2, 5, '6']))
print(my_any([0]))
print(my_any([]))
```
